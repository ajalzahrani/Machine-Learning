```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
```

```python
data = load_iris()
X = data.data
y = data.target
print(data.feature_names)
print(data.target_names)
print(X[:3], y[:3])
```

```python
plt.figure(figsize=(10,6))
plt.plot(X[y==0, 2], X[y==0, 3], "yo", label="Cluster 1")
plt.plot(X[y==1, 2], X[y==1, 3], "bs", label="Cluster 2")
plt.plot(X[y==2, 2], X[y==2, 3], "g^", label="Cluster 3")
plt.xlabel("Petal length")
plt.ylabel("Petal width")
plt.legend(loc="upper left")
plt.grid()
plt.show()
```

```python
X = X[:,2:]
```

```python
X[:5]
```

```python
neigh = KNeighborsClassifier(n_neighbors=5) # K = n_neighbors
neigh.fit(X,y)
y_pred = neigh.predict(X)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y, y_pred)
```

```python
from sklearn.inspection import DecisionBoundaryDisplay

plt.figure(figsize=(10,5))
disp = DecisionBoundaryDisplay.from_estimator(neigh,X,response_method="predict")
#plt.plot(X[:,0],X[:,1],'ko')
plt.plot(X[y==0, 0], X[y==0, 1], "yo", label="Class 1")
plt.plot(X[y==1, 0], X[y==1, 1], "bs", label="Class 2")
plt.plot(X[y==2, 0], X[y==2, 1], "g^", label="Class 3")

plt.show()
```

## Elbow Technique

```python
score = []
for k in [2,3,4,5,6,7,8,9,10]:
    neigh = KNeighborsClassifier(n_neighbors=k)
    neigh.fit(X,y)
    y_pred = neigh.predict(X)
    score.append(accuracy_score(y,y_pred))
    print('For the of K =',k,'the accuracy score is',accuracy_score(y,y_pred))

plt.plot([2,3,4,5,6,7,8,9,10], score)
plt.grid()
plt.show()
```

### For this Iris data with petal length and petal width as the features, the best value for K is '3'

```python
neigh = KNeighborsClassifier(n_neighbors=3) # K = n_neighbors
neigh.fit(X,y)
y_pred = neigh.predict(X)

print(accuracy_score(y, y_pred))

plt.figure(figsize=(10,5))
disp = DecisionBoundaryDisplay.from_estimator(neigh,X,response_method="predict")
#plt.plot(X[:,0],X[:,1],'ko')
plt.plot(X[y==0, 0], X[y==0, 1], "yo", label="Class 1")
plt.plot(X[y==1, 0], X[y==1, 1], "bs", label="Class 2")
plt.plot(X[y==2, 0], X[y==2, 1], "g^", label="Class 3")

plt.show()
```

```python
neigh = KNeighborsClassifier(n_neighbors=3, metric='euclidean' ) # K = n_neighbors
neigh.fit(X,y)
y_pred = neigh.predict(X)

print(accuracy_score(y, y_pred))

plt.figure(figsize=(10,5))
disp = DecisionBoundaryDisplay.from_estimator(neigh,X,response_method="predict")
#plt.plot(X[:,0],X[:,1],'ko')
plt.plot(X[y==0, 0], X[y==0, 1], "yo", label="Class 1")
plt.plot(X[y==1, 0], X[y==1, 1], "bs", label="Class 2")
plt.plot(X[y==2, 0], X[y==2, 1], "g^", label="Class 3")

plt.show()
```
