```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
m = 20
X = 3 * np.random.rand(m, 1)
y = 1 + 0.5 * X + np.random.randn(m, 1) / 1.5
X_new = np.linspace(0, 3, 100).reshape(100, 1)
```

```python
plt.figure(figsize=(8,6))
plt.plot(X, y, "bo", linewidth=3)
plt.xlabel("$x_1$", fontsize=18)
plt.axis([0, 3, 0, 3.5])
plt.show()
```

# Ridge

```python
from sklearn.linear_model import Ridge
```

```python
ridge_reg = Ridge(alpha=5, solver="cholesky", random_state=42)
ridge_reg.fit(X, y)
ridge_reg.predict([[1.5]])
```

```python
plt.figure(figsize=(8,6))
plt.plot(X, y, "bo", linewidth=4)
plt.xlabel("$x_1$", fontsize=18)
plt.axis([0, 3, 0, 4])
y_new_regul = ridge_reg.predict(X_new)
plt.plot(X_new, y_new_regul, 'r-')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)

ridge_reg = Ridge(alpha=5, solver="lsqr", random_state=42)
ridge_reg.fit(X, y)

plt.figure(figsize=(8,6))
plt.plot(X, y, "bo", linewidth=5)
plt.xlabel("$x_1$", fontsize=18)
plt.axis([0, 3, 0, 4])
y_new_regul = ridge_reg.predict(X_new)
plt.plot(X_new, y_new_regul, 'r-',label='Ridge')
plt.plot(X_new, lin_reg.predict(X_new), 'g-',label='Linear')
plt.legend(loc="upper left", fontsize=15)
plt.show()
```

```python
ridge_reg.intercept_, ridge_reg.coef_
```

# Lasso

```python
from sklearn.linear_model import Lasso
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X, y)

lasso_reg = Lasso(alpha=0.1)
lasso_reg.fit(X,y)

plt.figure(figsize=(8,6))
plt.plot(X, y, "bo", linewidth=3)
plt.xlabel("$x_1$", fontsize=18)
plt.axis([0, 3, 0, 4])
y_new_regul = lasso_reg.predict(X_new)
plt.plot(X_new, y_new_regul, 'r-',label='Lasso')
plt.plot(X_new, lin_reg.predict(X_new), 'g-',label='Linear')
plt.legend(loc="upper left", fontsize=15)
plt.show()
```

```python
lasso_reg.coef_, lasso_reg.intercept_
```

# Elastic-Net

```python
from sklearn.linear_model import ElasticNet
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X, y)

Elas_reg = ElasticNet(alpha=0.2, l1_ratio=0.1)
Elas_reg.fit(X, y)

plt.figure(figsize=(8,6))
plt.plot(X, y, "bo", linewidth=3)
plt.xlabel("$x_1$", fontsize=18)
plt.axis([0, 3, 0, 4])
y_new_regul = Elas_reg.predict(X_new)
plt.plot(X_new, y_new_regul, 'r-',label='Elastic-Net')
plt.plot(X_new, lin_reg.predict(X_new), 'g-',label='Linear')
plt.legend(loc="upper left", fontsize=15)
plt.show()
```
