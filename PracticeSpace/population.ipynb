```python
df = pd.read_csv('../1_Dataset/Population estimates by gender, nationality, and region 2010 - 2022_data.csv')
```

```python
df['Population estimates'] = df['Population estimates'].str.replace(',', '')
df['Population estimates'] = df['Population estimates'].astype(float)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")

# Assuming df is your DataFrame and it has columns 'Year', 'Region' and 'Population estimates'
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x="Year", y="Population estimates", hue="Region")

plt.title('Population estimates over time for each Region')
plt.show()
```
