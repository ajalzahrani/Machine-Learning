```python
import pandas as pd
import numpy as np

# Function to generate random categorical values
def random_categorical_values(categories, size):
    return np.random.choice(categories, size=size)

# Set seed for reproducibility
np.random.seed(42)

# Generate 100 rows of data
num_rows = 100

data = {
    'Age': np.random.randint(18, 40, size=num_rows),
    'Income': np.random.randint(40000, 90000, size=num_rows),
    'Education': random_categorical_values(['Bachelor', 'Master', 'PhD'], num_rows),
    'Gender': random_categorical_values(['Male', 'Female'], num_rows),
    'City': random_categorical_values(['Chicago', 'San Francisco', 'Los Angeles'], num_rows),
    'Productivity': np.random.randint(60, 90, size=num_rows),
    'Rating': np.random.uniform(2.0, 4.5, size=num_rows),
    'Experience': np.random.randint(3, 10, size=num_rows),
    'Department': random_categorical_values(['IT', 'Marketing', 'Finance', 'HR'], num_rows)
}

df = pd.DataFrame(data)

# Save the generated dataset
df.to_csv('generated_data.csv', index=False)
```

```python

# EDA Plots

# Pairplot to visualize relationships between numerical variables
sns.pairplot(df, height=2, aspect=1.5)
plt.suptitle("Pairplot of Numerical Variables")
plt.show()
```

```python

# Distribution of numerical variables
plt.figure(figsize=(12, 8))
for i, column in enumerate(df.select_dtypes(include=[np.number])):
    plt.subplot(3, 3, i + 1)
    sns.histplot(df[column], kde=True)
    plt.title(f'Distribution of {column}')
plt.tight_layout()
plt.show()
```

```python

# Boxplot to identify outliers
plt.figure(figsize=(12, 8))
for i, column in enumerate(df.select_dtypes(include=[np.number])):
    plt.subplot(3, 3, i + 1)
    sns.boxplot(x=df[column])
    plt.title(f'Boxplot of {column}')
plt.tight_layout()
plt.show()
```

```python

# Countplot for categorical variables
plt.figure(figsize=(12, 6))
for i, column in enumerate(df.select_dtypes(include=['object'])):
    plt.subplot(2, 2, i + 1)
    sns.countplot(x=df[column])
    plt.title(f'Countplot of {column}')
plt.tight_layout()
plt.show()
```
