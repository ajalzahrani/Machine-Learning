## Outlier

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Age', data=df, color='orange')
plt.title('Boxplot of Age')
plt.show()
```

```python
# Mean and Standard deviation technique 
import numpy as np

age_mean = df['Age'].mean()
age_std = df['Age'].std()
df['Age'] = np.where((df['Age'] > age_mean + 3 * age_std) | \
                     (df['Age'] < age_mean - 3 * age_std), np.nan, df['Age'])
df['Age']
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
df
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Age', data=df, color='orange')
plt.title('Boxplot of Age')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Income', data=df, color='orange')
plt.title('Boxplot of Age')
plt.show()
```

```python
quartiles = df['Income'].quantile([0.25, 0.75])
iqr = quartiles[0.75] - quartiles[0.25]
print(quartiles[0.75] , quartiles[0.25], iqr)
ul = quartiles[0.75] + 1.5*iqr
ll = quartiles[0.25] - 1.5*iqr
print(ul, ll)
outliers = df[(df['Income'] > ul) | (df['Income'] < ll)]['Income']
print(outliers)
df['Income'].replace(outliers[9], iqr, inplace=True)
df['Income']
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Productivity', data=df, color='orange')
plt.title('Boxplot of Productivity')
plt.show()
```

```python
# computer mean
mean = df['Productivity'].mean()

# compute standard deviation
std = df['Productivity'].std()

# find the data points that are 3 standard deviation away from mean
df['Productivity'] = np.where((df['Productivity'] > age_mean + 3 * age_std) | \
                              (df['Productivity'] < age_mean - 3 * age_std), np.nan, df['Productivity'])

# impute the outliers with median
df['Productivity'].fillna(df['Productivity'].median(), inplace=True)
df
```
